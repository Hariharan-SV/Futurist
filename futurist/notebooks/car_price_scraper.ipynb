# Set price using API for missing entries

- Maruti - rename to maruti-suzuki 
- Tata, Toyota, Hyundai, Mercedes-Benz,BMW - Add 2 entries consecutively to get car name with "-"
- Datsun, FIAT, Force One, Honda, Volkswagen - not in market


## Fix Tata, Toyota, Hyundai, Mercedes-Benz,BMW

```python
companies = ['Tata', 'Toyota', 'Hyundai', 'Mercedes-Benz', 'BMW']
for index, row in df_main.iterrows():
  name_list = row['name'].split(sep=" ")
  if name_list[0] not in companies:
    continue
  
  if row['present_price'] is not None:
    continue
  
  url = "https://www.zigwheels.com/newcars/"+name_list[0].lower()+"/"+name_list[1].lower()+"-"+name_list[2].lower()
  req = requests.get(url)
  soup = BeautifulSoup(req.text, "html.parser")
  price = soup.find("meta", itemprop="highPrice")
  if price is not None:
    price = price['content']
  df_main.at[index, 'present_price'] = price
  print(df_main.iloc[index]['name'], df_main.iloc[index]['present_price'])
```

```python
unknown_present_price = set()
for index, row in df_main.iterrows():
  if row['present_price'] is not None:
    continue
  unknown_present_price.add(row['name'])
print(len(unknown_present_price))
```

## Fix Maruti cars

```python
for index, row in df_main.iterrows():
  name_list = row['name'].split(sep=" ")
  if name_list[0] != "Maruti":
    continue
  url = "https://www.zigwheels.com/newcars/maruti-suzuki"+"/"+name_list[1].lower()
  req = requests.get(url)
  soup = BeautifulSoup(req.text, "html.parser")
  price = soup.find("meta", itemprop="highPrice")
  if price is not None:
    price = price['content']
  else:
    url = "https://www.zigwheels.com/newcars/maruti-suzuki"+"/"+name_list[1].lower()+"-"+name_list[2].lower()
    req = requests.get(url)
    soup = BeautifulSoup(req.text, "html.parser")
    price = soup.find("meta", itemprop="highPrice")
    if price is not None:
      price = price['content']
  df_main.at[index, 'present_price'] = price
  print(df_main.iloc[index]['name'], df_main.iloc[index]['present_price'])
```

```python
unknown_present_price = set()
for index, row in df_main.iterrows():
  if row['present_price'] is not None:
    continue
  unknown_present_price.add(row['name'])
print(len(unknown_present_price))
```
